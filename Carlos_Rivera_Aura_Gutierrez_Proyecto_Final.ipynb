### **3. Prueba t de Student**

```python
import scipy.stats as stats # Import the scipy.stats module

# 3. Prueba t de Student para comparar medias entre hombres y mujeres
hombres_notas = df[df['SEXO'] == 'Masculino']['DEFINITIVA']
mujeres_notas = df[df['SEXO'] == 'Femenino']['DEFINITIVA']

t_stat_sexo, p_value_sexo = stats.ttest_ind(hombres_notas, mujeres_notas)

# Resultados
(t_stat_sexo, p_value_sexo)
```

- **Estadstico t:** 1.62
- **Valor p:** 0.11

**Interpretacin:** La prueba t sugiere que no hay una diferencia estadsticamente significativa en las medias de las notas entre hombres y mujeres, ya que el valor p (0.11) es mayor que 0.05.

Aunque la media de los hombres es ligeramente superior a la de las mujeres, esta diferencia no es estadsticamente significativa.
